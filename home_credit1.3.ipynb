```python
appl_docs=appl[['APPL_ID','DOC2','DOC3','DOC4','DOC5','DOC6','DOC7','DOC8','DOC9','DOC10','DOC11','DOC12','DOC13','DOC14',
                'DOC15','DOC16','DOC17','DOC18','DOC19','DOC20','DOC21','C_CONTACT_ADDR','C_PMT_ADDR','C_WORK_ADDR','EMP_PHONE',
                'H_PHONE','M_PHONE','VALID_PHONE','W_PHONE','OWN_CAR','OWN_REALITY','R_CONTACT_ADDR','R_PMT_ADDR','R_WORK_ADDR',
               'TARGET']]
#creating a seperate dataframe which contains all the documents,addresses,own car, own reality, phones, email data
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
#importing libraries for apriori and associaion rules
```

```python
basket = (appl_docs
          .groupby(['APPL_ID','DOC2','DOC3','DOC4','DOC5','DOC6','DOC7','DOC8','DOC9','DOC10','DOC11','DOC12','DOC13','DOC14',
                'DOC15','DOC16','DOC17','DOC18','DOC19','DOC20','DOC21','C_CONTACT_ADDR','C_PMT_ADDR','C_WORK_ADDR','EMP_PHONE',
                'H_PHONE','M_PHONE','VALID_PHONE','W_PHONE','OWN_CAR','OWN_REALITY','R_CONTACT_ADDR','R_PMT_ADDR','R_WORK_ADDR',
               'TARGET'])
          .sum().unstack().reset_index().fillna(0)
          .set_index('APPL_ID'))
#grouped all the features in appl_docs data
```

```python
frequent_itemsets = apriori(basket, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
#displaying all the rules
```

Found 128 rules with support threshold value as 0.3

```python
appl=appl.drop(['M_PHONE','VALID_PHONE','DOC4','DOC5','DOC7','DOC10','DOC12','DOC17','DOC19','DOC20'],axis=1)
#dropped the above columns in the main dataframe, after inferential testing
```

```python
appl_docs=appl_docs.drop(['M_PHONE','VALID_PHONE','DOC4','DOC5','DOC7','DOC10','DOC12','DOC17','DOC19','DOC20'],axis=1)
#dropped the above columns in the docs dataframe, after inferential testing
```

```python
basket1 = (appl_docs
          .groupby(['APPL_ID','DOC2','DOC3','DOC6','DOC8','DOC9','DOC11','DOC13','DOC14',
                'DOC15','DOC16','DOC18','DOC21','C_CONTACT_ADDR','C_PMT_ADDR','C_WORK_ADDR','EMP_PHONE',
                'H_PHONE','W_PHONE','OWN_CAR','OWN_REALITY','R_CONTACT_ADDR','R_PMT_ADDR','R_WORK_ADDR',
               'TARGET'])
          .sum().unstack().reset_index().fillna(0)
          .set_index('APPL_ID'))
```

```python
frequent_itemsets = apriori(basket1, min_support=0.3, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
#Found 8 rules after dropping insignificant features
```
